Histogram Plot

```python
df.hist(bins=50,figsize=(16,12), color="orange")
```

The key insights drawn out from the histogram plots are as below :

The price of the houses is found to be more concentrated between the range 50000  to 75000.
The lotsize for the houses is condensed between 2500sq meters to 5000sq meters. 
The houses with 3 bedrooms have more count than houses with 2 and 4 bedrooms. 
It is observed that houses having 1 bathroom are quite higher than the houses having 2 or 3 bathrooms.
It can also be noticed that houses having 2 stories have higher count than the houses with 1, 3 and 4 stories.
The count of houses without garage place is perceived to be much bigger compared to houses with garage place. 

It is also spotted that the houses with these features(recreation rooms, full basement and air-conditioning) is fewer than the houses without them.
